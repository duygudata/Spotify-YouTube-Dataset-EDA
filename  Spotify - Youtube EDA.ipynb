```python
# Since I believe the columns containing descriptions and links will not be useful for the analysis, 
# I am removing them.

df = df.drop(['Unnamed: 0','Url_spotify', 'Uri','Url_youtube','Description','Title'], axis=1)
df.columns
```

```python
# I am filling the empty values in 'Like' and 'Comment' as zeros.
# I am removing the remaining empty rows.
```

###### DATA VISUALISATION

```python
df.groupby("Channel")[["Likes","Views","Comments"]].sum().sort_values(by="Likes", ascending=False).reset_index().head(10).plot(
    kind="bar",
    x="Channel",
    subplots=True,
    layout=(1,3),
    figsize=(15,5),
    title="Top 10 Channels by Like Count",
    rot=90
)
plt.show()
```

```python
df.groupby("Artist")[["Views","Stream"]].sum().sort_values(by="Views", ascending=False).reset_index().head(10).plot(
    kind="bar",
    x="Artist",
    subplots=True,
    layout=(1,2),
    figsize=(15,5),
    title= "Top 10 Artists by YouTube View Rate",
    rot=90
)
plt.show()
```

```python
df_artist = df.set_index("Artist")
df_artist = df_artist[["Danceability","Views","Likes","Album_type"]]

df_artist_grouped = df_artist.groupby('Artist')['Likes'].sum().reset_index()
df_artist_sorted = df_artist_grouped.sort_values(by='Likes', ascending=False)
top_10_artists = df_artist_sorted.head(10)

sns.barplot(x='Likes', y='Artist', data=top_10_artists)
plt.xlabel('Like Count')
plt.ylabel('Artist')
plt.title('Top 10 Artists with the Highest Like Count')
plt.show()
```

```python
df.groupby("Artist")[["Views","Stream"]].sum().sort_values(by="Stream", ascending=False).reset_index().head(10).plot(
    kind="bar",
    x="Artist",
    subplots=True,
    layout=(1,2),
    figsize=(15,5),
    title= "Top 10 Artists by Spotify Play Rate",
    rot=90
)
plt.show()
```

```python
df.groupby("Track")[["Views","Stream"]].sum().sort_values(by="Views", ascending=False).reset_index().head(10).plot(
    kind="bar",
    x="Track",
    subplots=True,
    layout=(1,2),
    figsize=(15,5),
    title= "Top 10 Tracks by YouTube View Rate",
    rot=90
)
plt.show()
```

```python
df_track = df.set_index("Track")
df_track = df_track[["Danceability","Views","Likes","Album_type"]]

df_track_grouped = df_track.groupby('Track')['Likes'].sum().reset_index()
df_track_sorted = df_track_grouped.sort_values(by='Likes', ascending=False)
top_10_tracks = df_track_sorted.head(10)

sns.barplot(x='Likes', y='Track', data=top_10_tracks)
plt.xlabel('Like Count')
plt.ylabel('Track')
plt.title('Top 10 Tracks with the Highest Like Count')
plt.show()
```

```python
df.groupby("Track")[["Views","Stream"]].sum().sort_values(by="Stream", ascending=False).reset_index().head(10).plot(
    kind="bar",
    x="Track",
    subplots=True,
    layout=(1,2),
    figsize=(15,5),
    title= "Top 10 Tracks by Spotify Play Rate",
    rot=90
)
plt.show()
```

```python
df.groupby(['Album_type']).count().plot(kind='pie', y='Track', autopct='%1.0f%%', 
                                             title='"Distribution by Album Genre"', 
                                             legend=False)
```

```python
df.groupby("Album_type")[["Likes","Views","Comments"]].sum().sort_values(by="Likes", ascending=False).reset_index().plot(
    kind="bar",
    x="Album_type",
    subplots=True,
    layout=(1,3),
    figsize=(15,5),
    title= "Likes-Views-Comments by Album Genre",
    rot=45
)
plt.show()
```

```python
def grafikler(df):
    numeric_columns= [col for col in df.columns if df[col].dtype != "O"]
    for i in numeric_columns:
        fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(20,4))
        
        sns.histplot(x=df[i],bins=10,ax=axes[0])
        axes[0].set_title(i)

        sns.boxplot(x=df[i],ax=axes[1])
        axes[1].set_title(i)

       
```

```python
grafikler(df)
```

```python
# Correlation Matrix Dataframe
df_corr = df[['Liveness','Danceability',
                       'Loudness',
                       'Speechiness',
                       'Valence',
                       'Tempo',
                       'Acousticness',
                       'Views',
                        'Likes',
                        'Comments',
                       'Stream']].dropna().corr().round(2)
```

```python
corr_matrix = df_corr.corr(numeric_only=True)
plt.figure(figsize=(10, 6))  
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',linewidths=.5,cbar_kws={"shrink": .9})  
plt.title("Correlation Matrix")
plt.show()
```
